```python
y = X['revenue']
X.drop(['revenue'], axis =1, inplace=True)
```

```python
X.replace({'campaign':{'campaign 1':1}},inplace=True)
test.replace({'campaign':{'campaign 1':1}},inplace=True)
```

```python
dic_to_replace = {"adgroup":{'adgroup 1':1,'adgroup 2':2,'adgroup 3':3,'adgroup 4':4}}
X.replace(dic_to_replace,inplace=True)
test.replace(dic_to_replace,inplace=True)
```

```python
X.campaign.astype(int)
test.campaign.astype(int)
```

```python
X.adgroup.astype(int)
test.adgroup.astype(int)
```

```python
X.ad =X.ad.str.slice(3).astype(int)
test.ad =test.ad.str.slice(3).astype(int)
```

```python
X['date']=pd.to_datetime(X.date)
test['date']=pd.to_datetime(test.date)
```

```python
X['year']=X.date.dt.year
test['year']=test.date.dt.year
```

```python
X['month']=X.date.dt.month
test['month']=test.date.dt.month
```

```python
X['day']=X.date.dt.day
test['day']=test.date.dt.day
```

```python
test.dtypes
```

```python
X.drop(['date'], axis =1, inplace=True)
test.drop(['date'], axis =1, inplace=True)
```

```python
test.head()
```

```python
X['ctr']=X.clicks/X.impressions
X['cpc']=X.cost/X.clicks
X['cpa']=X.cost/X.conversions
```

```python
test['ctr']=test.clicks/test.impressions
test['cpc']=test.cost/test.clicks
test['cpa']=test.cost/test.conversions
```

```python
X.fillna(0,inplace=True)
```

```python
test.fillna(0,inplace=True)
```

```python
X.replace([np.inf,-np.inf],0,inplace=True)
```

```python
test.replace([np.inf,-np.inf],0,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regr = DecisionTreeRegressor(criterion='mae',max_depth=5)
regr.fit(X_train, y_train)
```

```python
y_pred = regr.predict(X_valid)
```

```python
from sklearn.metrics import r2_score
print("R2 score for Training data: ",r2_score(y_valid, y_pred))
```
